```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ex=0.00001
E=[0.0,0.01]
k=8.61*1e-5
n=2
T=np.linspace(1,500,4000)
```

```python
def diff_(f,x):
    dx=0.001
    return (f(x+dx)-f(x))/dx
```

```python
def Z(T):
    z=0
    bt=1/(k*T)
    for i in range(len(E)):
            z+=np.exp(-bt*E[i])
    return z
```

```python
plt.plot(T,Z(T))
plt.title('Partition-Function')
plt.xlabel("Temp.(K)")
plt.ylabel("Z")
#plt.xlim(1,20)
plt.show()
```

```python
def F(T):
    return -n*k*T*np.log(Z(T))
```

```python
plt.plot(T,F(T))
plt.title('Hemholtz-Free-Eneregy')
plt.xlabel("Temp.(K)")
plt.ylabel("F")
plt.show()
```

```python
def S(T):
    return -diff_(F,T)
```

```python
plt.plot(T,S(T))
plt.title('Entropy')
plt.xlabel("Temp.(K)")
plt.ylabel("S")
plt.show()
```

```python
#function used in U below
def X(T):
    return np.log(Z(T)) 
```

```python

def U(T):
    return n*(k*T**2)*diff_(X,T)
```

```python
plt.plot(T,U(T))
plt.title('Average-Energy')
plt.xlabel("Temp.(K)")
plt.ylabel("U")
plt.show()
```

```python
def Cv(T):
    return diff_(U,T)
```

```python
plt.plot(T,Cv(T))
plt.title('Specific-Heat')
plt.xlabel("Temp.(K)")
plt.ylabel("Cv")
plt.xlim(0,400)
plt.show()
```
