# Exp.- Energy Distribution (MB,BE,FD)

# Statistical Mechanics 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n= 1000
k= 8.625*10e-5

#E-array
E = np.linspace(0.001,10,n)
E_ = np.linspace(4,10,n)

#main array
aa=[0,1,-1]
lba=['MB','FD','BE']
ls=['-',':','-.']
```

```python
def MB_F_B(u,T,E,a):
    return 1/(np.exp((E-u)/(k*T)) + a)
```

```python
def u_(u,a,E,nm,T,l):
    for i in range(len(u)):
        plt.plot(E,MB_F_B(u[i],T,E,a),label=nm+' at u = '+str(u[i]),ls=ls[i])
    l
    plt.grid()
    plt.legend(loc='best')
    plt.xlabel("Energy")
    plt.title(nm+' distribution at T = '+str(T)+" K")
    plt.show()
```

```python
def T_(T,a,E,nm,u,l):
    for i in range(len(T)):
        plt.plot(E,MB_F_B(u,T[i],E,a),label=nm+' at T = '+str(T[i])+" K",ls=ls[i])
    l
    plt.grid()
    plt.xlabel("Energy")
    plt.legend(loc='best')
    plt.title(nm+' distribution at u = '+str(u))
    plt.show()
```

```python
def a_(E,a,lba):
    t=0
    for i in a:
        plt. plot(E,MB_F_B(0,300.0,E,i),label=lba[t]+' distribution at u = '+str(0)+' & T= '+str(300.0)+'K',ls=ls[t])
        t+=1
    plt.legend(loc='best')
    plt.xlabel("Energy")
    plt.title( 'Distribution Function')
    plt.ylim([0,1])
    plt.xlim([0,1])
    plt.grid()
    plt.show()
```

```python
#all
a_(E,aa,lba)
```

```python
#Fd
u=[4,5,6]
l=plt.xlim(0,10)
u_(u,aa[1],E,lba[1],3.0,l)
```

```python
#Fd
T=[300.0,400.0,500.0]
l=plt.xlim(0,10)
T_(T,aa[1],E,lba[1],5,l)
```

```python
#BE
T=[300.0,400.0,500.0]
l=plt.ylim(0,0.0001)
T_(T,aa[2],E_,lba[2],1,l)
```
