# Bosons

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e=1.6e-19
k=1.38e-23
h=6.626e-34
v=1.0
m=4*1.66e-27
c=3e8   #m and c are replaced b/w relativistic and non relativistic
```

```python
#non relativistic
Cn=lambda s,u:(2*s+1)*(2*np.pi*v*(2*m)**1.5)/(h**3)
```

```python
#relativistic
Cr=lambda s,u:(2*s)*(4*np.pi*v)/(h**3*c**3)
```

```python
def BE(E,T,u,f):
    bt=1/(k*T)
    return 1/(np.exp(bt*f*(E-u))-1)
```

```python
def gE_n(E,s,u):
    return Cn(s,u)*(E)**0.5
```

```python
def gE_r(E,s,u):
    return Cr(s,u)*(E)**2
```

```python
def B_n(T,s,u,f):
    return gE_n(E,s,u)*BE(E,T,u,f)
```

```python
def B_r(T,s,u,f):
    return gE_r(E,s,u)*BE(E,T,u,f)
```

# Non-Relativistic Bosons

```python
T1=1000.0#K
T2=100000.0#K
s=1
u=-1.0#eV
f=e
E=np.linspace(0,0.5,1000)#in eV
```

```python
#density of states
plt.plot(E,gE_n(E,s,u))
plt.xlabel('Energy in eV')
plt.ylabel('g(E)')
plt.title("density of states")
plt.show()
```

```python
#BE distribution
plt.plot(E,BE(E,T1,u,f),label=str(T1)+'K')
plt.xlabel('Energy in eV')
plt.ylabel('FD')
plt.title('Bose Einstein distribution')
plt.legend(loc='best')
plt.show()
plt.plot(E,BE(E,T2,u,f),label=str(T2)+'K')
plt.xlabel('Energy in eV')
plt.ylabel('BE')
plt.title('Bose Einstein distribution')
plt.legend(loc='best')
plt.show()
```

```python
#density of particles
plt.plot(E,B_n(T1,s,u,f),label=str(T1)+'K')
plt.xlabel('Energy in eV')
plt.ylabel('dN/dE')
plt.title('Non Relativistic Bosons at u='+str(u)+'eV, s='+str(s))
plt.legend(loc='best')
plt.show()
plt.plot(E,B_n(T2,s,u,f),label=str(T2)+'K')
plt.xlabel('Energy in eV')
plt.ylabel('dN/dE')
plt.title('Non Relativistic Bosons at u='+str(u)+'eV, s='+str(s))
plt.legend(loc='best')
plt.show()
```

# Relativistic Bosons

```python
T1=10e10#K
T2=10e20#K
s=1
u=-1.0#MeV
E=np.linspace(0,6,1000)#in MeV
f=e*10e6
```

```python
#density of states
plt.plot(E,gE_r(E,s,u))
plt.xlabel('Energy in MeV')
plt.ylabel('g(E)')
plt.title("density of states")
plt.show()
```

```python
#BE distribution
plt.plot(E,BE(E,T1,u,f),label='10^10 K')
plt.xlabel('Energy in MeV')
plt.ylabel('BE')
plt.title('Bose Einstein distribution')
plt.legend(loc='best')
plt.show()
plt.plot(E,BE(E,T2,u,f),label='10^20 K')
plt.xlabel('Energy in MeV')
plt.ylabel('BE')
plt.title('Bose Einstein distribution')
plt.legend(loc='best')
plt.show()
```

```python
#density of particles
plt.plot(E,B_r(T1,s,u,f),label='10^10 K')
plt.xlabel('Energy in MeV')
plt.ylabel('dN/dE')
plt.title('Relativistic Bosons at u='+str(u)+'MeV, s='+str(s))
plt.legend(loc='best')
plt.show()
plt.plot(E,B_r(T2,s,u,f),label='10^20 K')
plt.xlabel('Energy in MeV')
plt.ylabel('dN/dE')
plt.title('Relativistic Bosons at u='+str(u)+'MeV, s='+str(s))
plt.legend(loc='best')
plt.show()
```
