```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k=1.38e-23
h=6.626e-34
c=2.99e8
```

```python
def P_v(v,T):
    return (8*np.pi*v**2*(h*v))/(c**3*np.exp((h*v)/(k*T))-1)
```

```python
def R_v(v,T):
    return ((8*np.pi*v**2)/(c**3))*k*T
```

```python
def P_l(l,T):
    return (8*np.pi*h*c)/((l**5)*(np.exp((h*c)/(k*T*l))-1))
```

```python
def R_l(l,T):
    return (8*np.pi*k*T)/(l**4)
```

```python
n=1000
V=np.linspace(0.001,10e14,n)
Tr=[500,650,800]
L=np.linspace(10e-7,40*10e-6,1000)
cr=['k','b','r']
```

```python
cx=0
for i in Tr:
    plt.plot(V,P_v(V,i),label=str(i)+'K',c=cr[cx])
    plt.plot(V,R_v(V,i),c=cr[cx])
    plt.xlim(0,0.2e15)
    plt.ylim(0.,4.5e-18)
    cx+=1
plt.grid(ls=':')
plt.legend(loc='best')
plt.title('Black-Body-Radiation rayleigh vs plank')
plt.xlabel('frequency(v)')
plt.ylabel("U(v)")
plt.show()
```

```python
cy=0
for i in Tr:
    plt.plot(L,P_l(L,i),label=str(i)+'K',c=cr[cy])
    plt.plot(L,R_l(L,i),c=cr[cy])
    plt.xlim(0.000001,0.00003)
    plt.ylim(0,61)
    cy+=1
plt.grid(ls=':')
plt.legend(loc='best')
plt.title('Black-Body-Radiation rayleigh vs plank')
plt.xlabel('wavelength($\lambda$)')
plt.ylabel("U($\lambda$)")
plt.show()
```

```python
#wein's law
R = np.linspace(1e-8,40e-6,n)
def wien(T):
    w = P_l(R,T)
    np.max(w)
    for i in R:
        if P_l(i,T)==np.max(w):
            return i*T
for i in Tr:
    print('mT: at',i,'K =',wien(i))
```
