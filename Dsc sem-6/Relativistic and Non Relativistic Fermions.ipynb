# Fermions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e=1.6e-19
k=1.38e-23
h=6.626e-34
v=1.0
m=9.1e-31
c=3e8   #m and c are replaced b/w relativistic and non relativistic
```

```python
#non relativistic
Cn=lambda s,u:(2*s+1)*(2*np.pi*v*(2*m)**1.5)/(h**3)
```

```python
#relativistic
Cr=lambda s,u:(2*s)*(4*np.pi*v)/(h**3*c**3)
```

```python
def FD(E,T,u,f):
    bt=1/(k*T)
    return 1/(np.exp(bt*f*(E-u))+1)
```

```python
def gE_n(E,s,u):
    return Cn(s,u)*(E)**0.5
```

```python
def gE_r(E,s,u):
    return Cr(s,u)*(E)**2
```

```python
def F_n(T,s,u,f):
    return gE_n(E,s,u)*FD(E,T,u,f)
```

```python
def F_r(T,s,u,f):
    return gE_r(E,s,u)*FD(E,T,u,f)
```

```python
ls=['-','-.']
```

# Non-Relativistic Fermions

```python
T=[100.0,1000.0]#K
s=0.5
u=1.0#eV
f=e
E=np.linspace(0,2,1000)#in eV
```

```python
#density of states
plt.plot(E,gE_n(E,s,u))
plt.xlabel('Energy in eV')
plt.ylabel('g(E)')
plt.title("density of states")
plt.show()
```

```python
#FD distribution
for i in range(2):
    plt.plot(E,FD(E,T[i],u,f),ls=ls[i],label=str(T[i])+'K')
    plt.xlabel('Energy in eV')
    plt.ylabel('FD')
plt.title('Fermi Dirac distribution')
plt.legend(loc='best')
plt.show()
```

```python
#density of Particles
for i in range(2):
    plt.plot(E,F_n(T[i],s,u,f),ls=ls[i],label=str(T[i])+'K')
    plt.xlabel('Energy in eV')
    plt.ylabel('dN/dE')
plt.title('Non Relativistic Fermions at u='+str(u)+'eV, s='+str(s))
plt.legend(loc='best')
plt.show()
```

# Relativistic Fermions



```python
T=[100,10e9]#K
s=0.5
u=1.0#MeV
E=np.linspace(0,2,1000)#in MeV
f=e*10e6
```

```python
#density of states
plt.plot(E,gE_r(E,s,u))
plt.xlabel('Energy in MeV')
plt.ylabel('g(E)')
plt.title("density of states")
plt.show()
```

```python
#FD distribution
for i in range(2):
    plt.plot(E,FD(E,T[i],u,f),ls=ls[i],label=str(T[i])+'K')
    plt.xlabel('Energy in MeV')
    plt.ylabel('FD')
plt.title('Fermi Dirac distribution')
plt.legend(loc='best')
plt.show()
```

```python
#density of Particles
for i in range(2):
    plt.plot(E,F_r(T[i],s,u,f),ls=ls[i],label=str(T[i])+'K')
    plt.xlabel('Energy in MeV')
    plt.ylabel('dN/dE')
plt.title('Relativistic Fermions at u='+str(u)+'MeV, s='+str(s))
plt.legend(loc='best')
plt.show()
```
